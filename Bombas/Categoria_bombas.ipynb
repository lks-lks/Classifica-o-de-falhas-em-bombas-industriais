```python
from datetime import datetime, timedelta
```

```python
def valida_datas(falhas_dict):
  todas_datas = []

  for tipo_falha, intervalos in falhas_dict.items():
    for inicio_str, fim_str in intervalos:
      inicio = datetime.strptime(inicio_str, '%Y-%m-%d %H:%M:%S')
      fim = datetime.strptime(fim_str, '%Y-%m-%d %H:%M:%S')
      todas_datas.append((inicio, fim))
  if not todas_datas:
    return None

  menor_data = min([inicio for inicio, _ in todas_datas])
  maior_data = max([fim for _, fim in todas_datas])

  duracoes = [(fim - inicio).total_seconds() for inicio, fim in todas_datas]
  media_duracao = timedelta(seconds=sum(duracoes) / len(duracoes))

  return menor_data, maior_data, media_duracao
```

```python
for i, falhas in enumerate([falhas_1, falhas_2, falhas_3], start=11):
  menor, maior, media = valida_datas(falhas)
  print(f'Dia {i}:')
  print(f'Menor data: {menor}')
  print(f'Maior data: {maior}')
  print(f'Média de duração: {media}')

#identifica as menores e maiore datas presentes no dicionário de cada dia.
```

#### Classificando as falhas e criando novos df com as colunas já classificadas



```python
def classifica_falhas(df, falhas_dict, col_data='Date'):
    """
    Classifica cada linha do DataFrame com base nos intervalos de falha.

    Parâmetros:
    - df: DataFrame que possui a coluna de tempo
    - falhas_dict: dicionário de falhas (ex: faults_day_12)
    - col_data: nome da coluna de tempo no DataFrame (default: 'Date')

    Retorna:
    - DataFrame com nova coluna 'label'
    """
    df = df.copy()
    df[col_data] = pd.to_datetime(df[col_data])

    df["label"] = "Unlabeled"

    # Para cada tipo de falha e seus intervalos
    for tipo, intervalos in falhas_dict.items():
        for inicio_str, fim_str in intervalos:
            inicio = pd.to_datetime(inicio_str)
            fim = pd.to_datetime(fim_str)
            mask = (df[col_data] >= inicio) & (df[col_data] <= fim)
            df.loc[mask, "label"] = tipo

    return df
```

### Modelo Random Forest

#### Preparativos para o modelo RF e pré processamento.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df_ml = df_filtrado[df_filtrado["LABEL"] != "Unlabeled"].copy()
df_ml["DATE"] = pd.to_datetime(df_ml["DATE"])
```

```python
le = LabelEncoder()
df_ml["LABEL_ENCODED"] = le.fit_transform(df_ml["LABEL"])
```

```python
features = [
    "SUCTION PRESSURE", "DISCHARGE PRESSURE", "FLUID TEMP",
    "TEMP_MB", "CURRENT", "VOLTAGE", "RPM_MOTOR",
    "V_X", "V_Z", "STATOR TEMPO"
]
```

```python
X = df_ml[features]
y = df_ml["LABEL_ENCODED"]
```

#### Separação treino/teste

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
#escalonar
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, stratify=y, random_state=42
)
#manter a mesma proporção
```

```python
print("Tamanho do conjunto de treino:", X_train.shape)
print("Tamanho do conjunto de teste:", X_test.shape)
```

#### Treinamento e avaliação do modelo

```python
model_rf = RandomForestClassifier(n_estimators=200, random_state=42)
model_rf.fit(X_train, y_train)
```

```python
y_pred = model_rf.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predito")
plt.ylabel("Real")
plt.title("Matriz de Confusão - Random Forest")
plt.tight_layout()
plt.show()
```

```python
print("\nRelatório de Classificação:\n")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```
